```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("D:/Datasets/world_data.csv")
data
```

```python
data.isnull().sum()
```

```python
data.columns
```

Clustering to find a relation between Popularity of a song and danceability.  

```python
plt.scatter(data["popularity"],data["danceability"])
plt.xlabel("popularity")
plt.ylabel("danceability")
```

```python
km = KMeans(n_clusters = 3)
y_predicted = km.fit_predict(data[["popularity","danceability"]])
y_predicted
```

```python
data["cluster"] = y_predicted
data.head()
```

```python
data["cluster"].nunique
```

```python
km.cluster_centers_
```

```python
data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
data3 = data[data.cluster==2]
plt.scatter(data1["popularity"],data1["danceability"],color = "green")
plt.scatter(data2["popularity"],data2["danceability"],color = "red")
plt.scatter(data3["popularity"],data3["danceability"],color = "blue")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = "purple",marker="*",label="centroid")
plt.xlabel("popularity")
plt.ylabel("danceability")
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(data[["popularity"]])
data["popularity"]=scaler.transform(data[["popularity"]])
scaler.fit(data[["danceability"]])
data["danceability"]=scaler.transform(data[["danceability"]])
```

```python
plt.scatter(data["popularity"],data["danceability"])
```

```python
km = KMeans(n_clusters = 3)
y_predicted = km.fit_predict(data[["popularity","danceability"]])
y_predicted
```

```python
data["cluster"] = y_predicted
data.head()
```

```python
km.cluster_centers_
```

```python
data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
data3 = data[data.cluster==2]
plt.scatter(data1["popularity"],data1["danceability"],color = "green")
plt.scatter(data2["popularity"],data2["danceability"],color = "red")
plt.scatter(data3["popularity"],data3["danceability"],color = "blue")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = "purple",marker="*",label="centroid")
plt.xlabel("popularity")
plt.ylabel("danceability")
plt.legend()
```

```python
el=[]
k_range = range(1,20)
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(data[["popularity","danceability"]])
    el.append(km.inertia_)
```

```python
 plt.plot(k_range,el)
```
