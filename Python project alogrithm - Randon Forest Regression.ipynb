```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("C:/Users/GOKUL/Downloads/diamonds.csv")
data
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data["cut"]=le.fit_transform(data["cut"])
data["color"]=le.fit_transform(data["color"])
data["clarity"]=le.fit_transform(data["clarity"])

data
```

```python
y = data["price"].values
y
```

```python
x= data.drop(["price",'Unnamed: 0'],axis = 1).values
x
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/4,random_state=4)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(n_estimators = 100,random_state = 5)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score  # to find the accuracy
score = r2_score(y_test,y_pred)
score
```

```python
y_predall = pd.DataFrame(y_pred)
y_predall
```

```python
newa = pd.concat([data,y_predall],axis=1)
newa
```

```python
newa.rename(columns={0:"Predictions"},inplace=True)
newa
```
