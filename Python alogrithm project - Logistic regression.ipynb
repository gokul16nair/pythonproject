```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/GOKUL/Downloads/weatherHistory.csv")
data
```

```python
data.columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data["Precip Type"]=le.fit_transform(data["Precip Type"])

data
```

```python
data["Precip Type"].nunique()
```

```python
y = data["Precip Type"].values
y
```

```python
x = data.iloc[:,4:11].values
x
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print (classification_report(y_test,y_pred))
```
