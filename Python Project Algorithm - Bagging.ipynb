```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:/Datasets/world_data.csv")
data
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
df = data.head(100)
df
```

```python
y = df["popularity"].values
y
```

```python
x = df.iloc[:,[5,6,8,10,11,12,13,14,15]].values
x
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .25,random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = svm.SVC()
nbg_clf = GaussianNB()
dec_clf = DecisionTreeClassifier()
```

```python
for clf in(log_clf,nbg_clf,svm_clf,rnd_clf,dec_clf):
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    print(clf.__class__.__name__,accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bagg_nb_clf = BaggingClassifier(LogisticRegression(),n_estimators = 500, max_samples = 10, bootstrap = True, n_jobs = 1, oob_score = True, max_features = 0.7)
bagg_nb_clf.fit(x_train,y_train)
y_pred0 = bagg_nb_clf.predict(x_test)
print("Accuracy by using bagging logistic is : ",accuracy_score(y_test,y_pred0))
```

```python
bagg_nb_clf = BaggingClassifier(GaussianNB(),n_estimators = 500, max_samples = 10, bootstrap = True, n_jobs = 1, oob_score = True, max_features = 0.7)
bagg_nb_clf.fit(x_train,y_train)
y_pred0 = bagg_nb_clf.predict(x_test)
print("Accuracy by using bagging GNB is : ",accuracy_score(y_test,y_pred0))
```

```python
bagg_nb_clf = BaggingClassifier(DecisionTreeClassifier(),n_estimators = 500, max_samples = 10, bootstrap = True, n_jobs = 1, oob_score = True, max_features = 0.7)
bagg_nb_clf.fit(x_train,y_train)
y_pred0 = bagg_nb_clf.predict(x_test)
print("Accuracy by using bagging decision tree is : ",accuracy_score(y_test,y_pred0))
```

```python
bagg_nb_clf = BaggingClassifier(RandomForestClassifier(),n_estimators = 500, max_samples = 10, bootstrap = True, n_jobs = 1, oob_score = True, max_features = 0.7)
bagg_nb_clf.fit(x_train,y_train)
y_pred0 = bagg_nb_clf.predict(x_test)
print("Accuracy by using bagging random forest classifier is : ",accuracy_score(y_test,y_pred0))
```

```python
bagg_nb_clf = BaggingClassifier(svm.SVC(),n_estimators = 500, max_samples = 10, bootstrap = True, n_jobs = 1, oob_score = True, max_features = 0.7)
bagg_nb_clf.fit(x_train,y_train)
y_pred0 = bagg_nb_clf.predict(x_test)
print("Accuracy by using bagging svm is : ",accuracy_score(y_test,y_pred0))
```
