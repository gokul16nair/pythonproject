```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = pd.read_csv("C:/Users/GOKUL/Downloads/diamonds.csv")
data1
```

```python
## The aim is to predict the prices of diamond based on different parameters
```

```python
data1.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data1["cut"]=le.fit_transform(data1["cut"])
data1["color"]=le.fit_transform(data1["color"])
data1["clarity"]=le.fit_transform(data1["clarity"])

data1
```

```python
y1 = data1["price"].values
y1
```

```python
x1 = data1.drop(["price"],axis = 1).values
x1
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x1 = sc.fit_transform(x1)
x1
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size = 0.33,random_state = 4)
```

```python
from sklearn.tree import DecisionTreeRegressor
clf = DecisionTreeRegressor(random_state = 1)
clf.fit(x1_train,y1_train)
```

```python
y1_pred = clf.predict(x1_test)
y1_pred
```

```python
df=pd.DataFrame(y1_pred)
df
```

```python
datanew=pd.concat([data1,df],axis=1)
datanew
```

```python
datanew.rename(columns={0:"Predictions"},inplace=True)
datanew
```

```python
from sklearn.metrics import r2_score  # to find the accuracy
score=r2_score(y1_test,y1_pred)
score
```
