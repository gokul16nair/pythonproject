```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("C:/Users/GOKUL/Downloads/Pumpkin_Seeds_Dataset.xlsx")
data
```

The objective is to predict the class of seeds

```python
df = data["Class"]
df
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["Class"]=le.fit_transform(data["Class"])
data
```

```python
y = data["Class"].values
y
```

```python
x = data.drop(["Class"],axis = 1).values
x
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=4)
```

```python
from sklearn.svm import SVC  
clf = SVC(kernel = "linear")
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```

```python
y_predall = pd.DataFrame(y_pred)
y_predall
```

```python
newa = pd.concat([data,y_predall,df],axis=1)
newa
```

```python
newa.rename(columns={0:"Predictions"},inplace=True)
newa
```
