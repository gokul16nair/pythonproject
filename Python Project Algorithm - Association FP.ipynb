```python
data_new = pd.DataFrame(data.groupby("InvoiceDate")["Description"].nunique().index) ## To arrange by ascending order date
data_new
```

```python
data_new["member_count"]=data.groupby("InvoiceDate")["CustomerID"].nunique().values ## show member count
data_new["Items_count"]=data.groupby("InvoiceDate")["Description"].nunique().values ## number count
data_new["Items"]=data.groupby("InvoiceDate")["Description"].unique().values ## List not number
data_new
```

```python
transactions = data_new["Items"]
transactions.head(3)
```

```python
freq_patt = pyfpgrowth.find_frequent_patterns(transactions=transactions,support_threshold = 5)
freq_patt
```
